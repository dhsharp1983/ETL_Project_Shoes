```python
#Import Dependencies
import pandas as pd
```

```python
#Create reference to CSV file
csv_path1 = "Mens_Shoes_df.csv"

#Import the CSV into a pandas DataFrame
Mens_Shoes_df = pd.read_csv(csv_path1, low_memory = False)
```

```python
#Create reference to CSV file
csv_path2 = "Womens_Shoes_df.csv"

#Import the CSV into a pandas DataFrame
Womens_Shoes_df = pd.read_csv(csv_path2, low_memory = False)
```

#### PRINT BOTH DATAFRAMES

```python
#Print the Mens dataframe
print("The Men's dataframe looks like...")
Mens_Shoes_df
```

```python
#Print the Womens dataframe
print("The Women's dataframe looks like...")
Womens_Shoes_df
```

#### MERGE BOTH DATAFRAMES

```python
#Combine the 2 dataframes
Dafiniti_Shoes_df = Mens_Shoes_df.append(Womens_Shoes_df)

#Show the 'shape' of the dataframe
shape = Dafiniti_Shoes_df.shape
print(f"The dataframes 'shape' is represented as: {shape}")
```

```python
#Print the dataframe
#Dafiniti_Shoes_df
```

#### RE-ORDER THE DATAFRAME

```python
#Convert prices column to Float 
Dafiniti_Shoes_df['Price'] = pd.to_numeric(Dafiniti_Shoes_df['Price($)'], errors = 'coerce')

#Sort DF based on Price
Dafiniti_Shoes_df = Dafiniti_Shoes_df.sort_values('Price($)', ascending=False)
```

```python
#Print the dataframe
#Dafiniti_Shoes_df
```

#### DROP AN INDEX COLUMN AND SET ID AS INDEX

```python
#Drop index column (because there is now 2)
Dafiniti_Shoes_df = Dafiniti_Shoes_df.drop(['Index'], axis = 1)

#Re-correct the given index or set 'ID' as index
Dafiniti_Shoes_df = Dafiniti_Shoes_df.set_index(["ID"])

#Show the 'shape' of the dataframe
shape = Dafiniti_Shoes_df.shape
print(f"The dataframes 'shape' is represented as: {shape}")
```

#### PRINT THE RESULTING DATAFRAME

```python
#Print the Dafiniti dataframe
print("The Dafiniti Shoe's dataframe looks like...")
Dafiniti_Shoes_df
```

#### SAVE THE DATAFRAME

```python
#Save dataframe as csv file
Dafiniti_Shoes_df.to_csv('Dafiniti_Shoes_df.csv') 
```
