```python
#Select only columns we need
women_df = women_df[["id", "prices.dateSeen", "brand", "prices.amountMax", "colors", "name"]]
women_df.head()
```

```python
#Replace "," with "/", so we can split strings all together
color_list = women_df.colors.str.replace(',', '/')

#Loop through the list and pick out the first color before "/"
color=[]
for x in color_list:
    color.append(x.split("/")[0])

#Create a new column with the simplified color
women_df["Color"] = color
```

```python
#Drop off rows with no brand
men_df = men_df.dropna(subset=['brand'])
women_df = women_df.dropna(subset=['brand'])

#Convert brand from each dataset to a list
brand_name_w = women_df["brand"].tolist()
brand_name_m = men_df["brand"].tolist()

#combine the list into one
brand_name = brand_name_w + brand_name_m

#Drop off duplicated brand names
from collections import OrderedDict
brand_name = list(OrderedDict.fromkeys(brand_name))
brand_name_sorted = brand_name.sort()

#Create an Brand_ID list
ID=range(1,len(brand_name)+1)
brand_name_ID=[]
for x in ID:
    brand_name_ID.append("BN"+str(x))

#Create an Brand_Name dataframe
Brand_Name_df = pd.DataFrame({"Brand_ID": brand_name_ID, 
                              "Brand_Name" :brand_name}, index=None)
```

```python
#===============================CREATE COLOR TABLE ========================
men_df1=men_df.copy()
women_df1=women_df.copy()

#Drop off rows with no brand
men_df1 = men_df1.dropna(subset=['Color'])
women_df1 = women_df1.dropna(subset=['Color'])

#Convert brand from each dataset to a list
color_w = women_df1["Color"].tolist()
color_m = men_df1["Color"].tolist()

#combine the list into one
combined_color = color_w + color_m

#Drop off duplicated brand names
from collections import OrderedDict
combined_color = list(OrderedDict.fromkeys(combined_color))

#Create an Brand_ID list
ID=range(1,len(combined_color)+1)
Color_ID=[]
for x in ID:
    Color_ID.append("CL"+str(x))

#Create an Brand_Name dataframe
Color_df = pd.DataFrame({"Color_ID": Color_ID, 
                              "Colors" :combined_color}, index=None)
```
