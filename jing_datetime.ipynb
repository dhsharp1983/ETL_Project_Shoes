```python
import pandas as pd
import datetime
```

```python
w_file = "Datafiniti_Womens_Shoes_Jun19.csv"
women_df = pd.read_csv(w_file)
women_df.head()
```

```python
#Select only columns we need
women_df = women_df[["id", "prices.dateSeen", "brand", "prices.amountMax", "colors", "name"]]
women_df.head()
len(women_df)
```

```python
#Filter out non-DateTime valued rows, and set value to "NaN"
women_df['prices.dateSeen'] = pd.to_datetime(women_df['prices.dateSeen'], errors='coerce')

#Drop all "NaN" valued rows in prices.dateSeen column
women_df = women_df.dropna(subset=['prices.dateSeen'])

#Convert the price.dateSeen column to string (from TimeStamp due to previous function to_datetime used)
women_df['prices.dateSeen']=women_df['prices.dateSeen'].astype(str)

#Convert column to list
dt_list = women_df["prices.dateSeen"].tolist()

#Clean up the list to only show yyyy-mm-dd formate
w_date = []
for x in dt_list:
    w_date.append(x.split(" ")[0])

#Add the formated list into the dataframe as new column "Date"
women_df["Date"] = w_date

#Drop off the price.seenDate column
women_df = women_df[["id", "brand", "Date", "prices.amountMax" , "colors", "name"]]
```
