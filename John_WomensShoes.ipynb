# CSV: Datafiniti_Womens_Shoes_Jun19

#### ===================================================================================================================

```python
#Import Dependencies
import pandas as pd
```

```python
#Create reference to CSV file
csv_path = "Datafiniti_Womens_Shoes_Jun19.csv"

#Import the CSV into a pandas DataFrame
Womens_Shoes_df = pd.read_csv(csv_path, low_memory = False)
Womens_Shoes_df
```

```python
#Show the 'shape' of the dataframe 
shape = Womens_Shoes_df.shape
print(f"The dataframes 'shape' is represented as: {shape}")
```

###### COLLECT ONLY REQUIRED COLUMNS

```python
#Select only columns we need
Womens_Shoes_df = Womens_Shoes_df[["id", "brand", "colors", "name", "prices.amountMax",  "prices.currency", "prices.dateSeen", 
                               "prices.isSale"]]
Womens_Shoes_df
```

```python
#Show the 'shape' of the dataframe
shape = Womens_Shoes_df.shape
print(f"The dataframes 'shape' is represented as: {shape}")
```

###### REMOVE ALL ROWS WITH NO DATA

```python
#Filter out those rows which does not contain any data - just in case
Womens_Shoes_df = Womens_Shoes_df.dropna() 

#Confirm duplicates are dropped
shape = Womens_Shoes_df.shape
print(f"The dataframes 'shape' is represented as: {shape}")
```

###### REMOVE ALL ROWS WHICH CONTAIN A CURRENCY OTHER THAN USD

```python
#Get names of indexes for which column Currency has value other than USD
Womens_Shoes_df_indexnames = (Womens_Shoes_df[Womens_Shoes_df['prices.currency'] != "USD"]).index
#Delete these row indexes from dataframe
Womens_Shoes_df.drop(Womens_Shoes_df_indexnames, inplace=True)

#Print the dataframe
Womens_Shoes_df
```

```python
#Show the 'shape' of the dataframe
shape = Womens_Shoes_df.shape
print(f"The dataframes 'shape' is represented as: {shape}")
```

###### REMOVE ALL ROWS WITH FALSE IN THE "prices.isSale" COLUMN

```python
#Get names of indexes for which column Currency has value other than USD
Womens_Shoes_df_indexnames = (Womens_Shoes_df[Womens_Shoes_df['prices.isSale'] == "FALSE"]).index
#Delete these row indexes from dataFrame
Womens_Shoes_df.drop(Womens_Shoes_df_indexnames, inplace=True)

#Print the dataframe
Womens_Shoes_df
```

```python
#Drop the 'prices.isSale' column as it is now useless data - was used to siphon the non-Sale prices.
Womens_Shoes_df = Womens_Shoes_df.drop(['prices.isSale'], axis = 1)
Womens_Shoes_df
```

```python
#Show the 'shape' of the dataframe
shape = Womens_Shoes_df.shape
print(f"The dataframes 'shape' is represented as: {shape}")
```

###### RENAME COLUMNS AND CORRECT THE INDEX FOR VISUAL PURPOSES

```python
#The original dataframe had an index, so re-arrange the index 
Womens_Shoes_df.index = [x for x in range(1, len(Womens_Shoes_df.values)+1)]
Womens_Shoes_df.index.name = 'Index'
Womens_Shoes_df
```

```python
#Rename the columns for visual purposes
Womens_Shoes_df = Womens_Shoes_df.rename(columns = {"id":"ID", "brand":"Brand", "colors":"Colours", "name":"Name", 
                                                "prices.amountMax":"Price", "prices.currency":"Currency", 
                                                "prices.dateSeen":"Date"})

#Reorganise the columns for visualisation
Womens_Shoes_df = Womens_Shoes_df[['ID', 'Name', 'Brand', 'Colours', 'Price', 'Currency', 'Date']]

#Print the dataframe
Womens_Shoes_df
```

```python
#Show the 'shape' of the dataframe
shape = Womens_Shoes_df.shape
print(f"The dataframes 'shape' is represented as: {shape}")
```
